```python
import pandas as pd
import numpy as np
```

```python
#loading training and test data
train_data=pd.read_csv(r'D:\Others\Downloads\\risk_analytics_train.csv',header=0)
test_data=pd.read_csv(r'D:\Others\Downloads\\risk_analytics_test.csv',header=0)
```

**Preprocessing the training dataset**

```python
print(train_data.shape)    #gives us the number of rows and columns in the dataset

train_data.head()          #gives us the 1st 6 entries in the dataset
```

```python
#finding the missing values
print(train_data.isnull().sum())    #finds and sums the number of NA values in the dataset 
#print(train_data.shape)
```

```python
#imputing categorical missing data with mode value

colname1=["Gender","Married","Dependents","Self_Employed", "Loan_Amount_Term"]

for x in colname1:
    train_data[x].fillna(train_data[x].mode()[0],inplace=True)
```

```python
train_data["Gender"].value_counts()
```

```python
print(train_data.isnull().sum())
```

```python
#imputing numerical missing data with mean value
train_data["LoanAmount"].fillna(train_data["LoanAmount"].mean(),inplace=True)
print(train_data.isnull().sum())
```

```python
#imputing values for credit_history column differently
train_data['Credit_History'].fillna(value=0, inplace=True)
#train_data['Credit_History']=train_data['Credit_History'].fillna(value=0)
print(train_data.isnull().sum())
```

```python
#transforming categorical data to numerical
from sklearn import preprocessing
colname=['Gender','Married','Education','Self_Employed','Property_Area',
         'Loan_Status']

le={}

for x in colname:
     le[x]=preprocessing.LabelEncoder()

for x in colname:
     train_data[x]=le[x].fit_transform(train_data[x])
        
#converted Loan status as Y-->1 and N-->0
```

# Creating training and testing datasets and running the model

```python
X_train=train_data.values[:,1:-1]    #[all rows:all columns starting from 1 and excluding the last column(i.e.,-1)]
Y_train=train_data.values[:,-1]      #[all rows:all columns starting from 0 and exclding last column]
Y_train=Y_train.astype(int)
```

```python
#test_data.head()
X_test=test_data.values[:,1:]
```

```python
from sklearn.preprocessing import StandardScaler
#use "fit" only on training data and not on testing data
#"transform" should be applied to both training and testing data

scaler=StandardScaler()

scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn import svm

svc_model=svm.SVC(kernel='rbf',C=1.0,gamma=0.1)
#from sklearn.linear_model import logistic regression
#svc_model= LogisticRegression()
svc_model.fit(X_train, Y_train)
Y_pred=svc_model.predict(X_test)
print(list(Y_pred))
```

```python
Y_pred_col=list(Y_pred)
#print(Y_pred_col)
```

```python
test_data=pd.read_csv(r'D:\Others\Downloads\\risk_analytics_test.csv',header=0)
test_data["Y_predictions"]=Y_pred_col
test_data.head()
```

```python
test_data.to_csv(r'D:\Others\Downloads\\risk_analytics_test.csv')    #to make the new file in the desired folder 
```

```python
#Using cross validation

from sklearn.linear_model import LogisticRegression
classifier=(LogisticRegression())
#classifier=svm.SVC(kernel='rbf',C=1.0,gamma=0.1)

#performing kfold_cross_validation
from sklearn.model_selection import KFold
kfold_cv=KFold(n_splits=10)
print(kfold_cv)

from sklearn.model_selection import cross_val_score
#running the model using scoring metric as accuracy
kfold_cv_result=cross_val_score(estimator=classifier,X=X_train,
y=Y_train, cv=kfold_cv)
print(kfold_cv_result)
#finding the mean
print(kfold_cv_result.mean())


for train_value, test_value in kfold_cv.split(X_train):
    classifier.fit(X_train[train_value], Y_train[train_value]).predict(X_train[test_value])


Y_pred=classifier.predict(X_test)
#print(list(zip(Y_test,Y_pred)))
```
