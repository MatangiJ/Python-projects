```python
#discretization & Binning (converting numerical to categorical data)
PassengerAge=titanic_df['Age']
PassengerAge=PassengerAge.dropna()        #drops the missing values in the age columns
Bins=[0,15,21,60,PassengerAge.max()]
Binlabels=['Children','Adolescents','Adult','Senior']
categories=pd.cut(PassengerAge,Bins,labels=Binlabels,include_lowest=True)
print(categories.value_counts())
```

```python
categories1=pd.cut(PassengerAge,Bins,include_lowest=True)
print(categories1.value_counts())
```

```python
new_df=pd.concat([PassengerAge,categories],axis=1)
new_df.head(10)                                    #observations(axis=1) & variables(axis=0) #merge for merging two files(both rows and columns)
```
