```python
import numpy as np
import pandas as pd
#%%
pd.set_option('display.max_columns',None)

#%%loading Data from excel file
data_train=pd.read_excel(r'C:\Users\Admin\Desktop\project\python project\Train_Data.xlsx', header=0)
data_test=pd.read_excel(r'C:\Users\Admin\Desktop\project\python project\Test_Data.xlsx', header=0)

#%% Missing values for training data
data_train.isnull().sum()
#%%Missing values for testing data
data_test.isnull().sum()
#%%deleting unwanted column in train data
del data_train['issue_d']
del data_train['application_type']
#%% deleting unwanted columns in test
del data_test['issue_d']
del data_test['application_type']
#%%% size of data sets
print(data_train.shape)
data_test.shape
#%% filling Missing value for train data
data_train['tot_coll_amt'].value_counts()
data_train['tot_coll_amt'].fillna(value=0, inplace=True)
data_train['tot_coll_amt'].isnull().sum()
#%% 
data_train['tot_cur_bal'].isnull().sum()
data_train['tot_cur_bal'].fillna(value=0, inplace=True)
data_train['tot_cur_bal'].isnull().sum()
#%%
data_train['emp_length'].value_counts()
data_train['emp_length']=data_train['emp_length'].replace(['n/a'],np.nan)
data_train['emp_length'].isnull().sum()
data_train['emp_length'].fillna(data_train['emp_length'].mode()[0],inplace=True)
data_train['emp_length'].isnull().sum()
#%%
data_train['revol_util'].fillna(data_train['revol_util'].mean(), inplace=True)
data_train['revol_util'].isnull().sum()
#%%
data_train['collections_12_mths_ex_med'].fillna(data_train['collections_12_mths_ex_med'].mean(),inplace= True)
data_train['collections_12_mths_ex_med'].isnull().sum()
#%%% Filling Missing values for Test data
data_test['revol_util'].fillna(data_test['revol_util'].mean(),inplace=True)
data_test['revol_util'].isnull().sum()
#%%%
data_test['emp_length'].fillna(data_test['emp_length'].mode()[0],inplace=True)
data_test['emp_length'].isnull().sum()
#%%% Converting all categorical Variables into Numeric for train data
data_train.dtypes
colname=['term','grade','emp_length','home_ownership','verification_status','pymnt_plan']
from sklearn import preprocessing
le={}
for x in colname:
    le[x]=preprocessing.LabelEncoder()
for x in colname:
    data_train[x]=le[x].fit_transform(data_train[x])

data_train.dtypes
#%% converting all categorical data into numeric for test data
data_test.dtypes
from sklearn import preprocessing
le1={}
for x in colname:
    le1[x]=preprocessing.LabelEncoder()
for x in colname:
    data_test[x]=le1[x].fit_transform(data_test[x])

data_test.dtypes
#%%% creating X and Y
X_train=data_train.values[:,1:-1]
Y_train=data_train.values[:,-1]
X_test=data_test.values[:,1:-1]
Y_test=data_test.values[:,-1]
#%% Model buliding
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
#%%%% to Predict the test data
Ypred=classifier.predict(X_test)
print(list(zip(Y_test,Ypred)))
#%%%% to check accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cfm=confusion_matrix(Y_test,Ypred)
acc= accuracy_score(Y_test,Ypred)
print('classification report: ',acc)
print(classification_report(Y_test,Ypred))
#%% decision tree
from sklearn.tree import DecisionTreeClassifier 
model_DT=DecisionTreeClassifier(criterion='gini',random_state=10)
model_DT.fit(X_train,Y_train)
#%%% Predict Y for test
Ypred1=model_DT.predict(X_test)
#%%%% to check accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cfm1=confusion_matrix(Y_test,Ypred)
acc1= accuracy_score(Y_test,Ypred)
print('classification report: ',acc1)
print(classification_report(Y_test,Ypred))
#%%%%%


```
